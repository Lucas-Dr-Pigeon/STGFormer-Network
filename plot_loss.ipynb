```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
loss_log = pd.read_csv('./logs/loss_log.csv')
```

```python
baseline_loss_log = pd.read_csv('./logs/baseline_loss_log.csv')
```

```python
plt.plot(loss_log['epoch'], loss_log['train_loss'], color='red', label='TASTformer-train-loss')
plt.plot(loss_log['epoch'], loss_log['val_loss'], color='blue', label='TASTformer-val-loss')
# plt.plot(baseline_loss_log['epoch'], baseline_loss_log['train_loss'], color='red', label='PropSTGformer-train-loss', linestyle='dashed')
# plt.plot(baseline_loss_log['epoch'], baseline_loss_log['val_loss'], color='blue', label='PropSTGformer-val-loss', linestyle='dashed')
plt.xlim(0, 80)
plt.grid()
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend()
```
